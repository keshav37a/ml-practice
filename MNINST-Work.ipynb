```python
dataset_train = pd.read_csv("C:/Users/Keshav/Desktop/ML 2019/machine-learning-june-2019/Dataset/MNIST/train.csv")
dataset_test = pd.read_csv("C:/Users/Keshav/Desktop/ML 2019/machine-learning-june-2019/Dataset/MNIST/test.csv")
```

```python
data_train=dataset_train.values
data_test=dataset_test.values
print(type(data_train))
print(type(data_test))
```

```python
x_train_values=data_train[:,1:]
y_train_values=data_train[:,0]
print(type(y_train_values))
print(x_train_values.shape)
print(y_train_values.shape)
print(data_test.shape)
print(y_train_values[0:10,])
```

```python
def calc_dist(p1, p2):
    return np.sum((p2-p1)**2)**0.5
```

```python
def knn(X, Y, test, k=10):
    m=X.shape[0]
    d=[]
    print(m)
    print(X.shape)
    print(test.shape)
    for i in range(m):
        dist=calc_dist(X[i], test)
        d.append((dist, Y[i]))
    d=np.array(sorted(d))[:,1]
    d=d[:k]
    t =  np.unique(d,return_counts=True)
    idx = np.argmax(t[1])
    pred = int(t[0][idx])
    print("The prediction is " +str(pred))
    plt.imshow(test.reshape((28,28)),cmap='gray')
    plt.show()
    
```

```python
for i in range(100):    
    knn(x_train_values, y_train_values, data_test[i])
    
```
